#### Def price value

If the value within the “prices” column within the given row is less than or equal to 5, then return the string “Low-range product.”

Or else, if the value within the “prices” column within the given row is greater than 5 and less than or equal to 15, then return the string “Mid-range product.”

Or else, if the value of the “prices” column within the given row is greater than 15, then return the string “High-range product.”

Or else, return the string “Not enough data.”

```python
# Apply the function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

#### Q2: Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
updated_result = []

for value in df_orders_products_merged["orders_day_of_week"]:
  if value == 0:
    updated_result.append("Busiest days")
  elif value == 1:
    updated_result.append("Busiest days")
  elif value == 3:
    updated_result.append("Slowest days")
  elif value == 4:
    updated_result.append("Slowest days")
  else:
    updated_result.append("Regularly busy")
```

#### Q4: When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
df_orders_products_merged['orders_time_of_day'].value_counts(dropna = False)
```

```python
# Create a new column

new_result = []

for value in df_orders_products_merged["orders_time_of_day"]:
  if value <=5:
    new_result.append("Fewest_orders")
  elif value >= 6 and value <=8 or value >= 18 and value <=24:
    new_result.append("Average_orders")
  elif value >= 9 and value <=17:
    new_result.append("Most_orders")
  else:
    new_result.append("Not sure")
```

```python
new_result
```

```python
#Create new column

df_orders_products_merged['busiest_period_of_day'] = new_result
```
